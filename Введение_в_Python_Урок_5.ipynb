```python
from decimal import Decimal

a = Decimal("0.2") - Decimal("0.1") - Decimal("0.1")
print(a)
```

```python
# Модуль fractions может быть использован для осуществления математических
# операций с дробями. Например:
```

```python
from fractions import Fraction

b = Fraction(6, 4)
print(b)
c = Fraction(7, 12)
print(c)
d = b * c
print(d.numerator, d.denominator)
print(float(d))

x = 5.75
y = Fraction(*x.as_integer_ratio())
print(y)
```

```python
# Случайные числа, модуль random
```

```python
import random

print(random.random())
print(random.randint(10, 50))

a_list = ["Привет", "Здравствуйте", "Hello"]
random.shuffle(a_list)
print(a_list)

print(random.choice(a_list))
print(random.randrange(10, 50, 4))

b_list = [i for i in range(100)]
random.shuffle(b_list)
print(b_list)
print(random.sample(b_list, 5))
```

```python
# Основы работы с numpy (массивы)
```

```python
import sys

import numpy as np


a_list = [i for i in range(50)]
b_list = [i for i in range(50)]

# создание массивов из списков
a0 = np.array(a_list)
b0 = np.array(b_list)


if __name__ == '__main__':
    # Сложение массивов
    c0 = a0 + b0
    print(c0)

    # Умножение массивов
    c0 = a0 * b0
    print(c0)

    # Сложение массивов методом arange
    a1 = np.arange(0, 2*np.pi, np.pi/6)
    # Применение функции sin к массиву
    s1 = np.sin(a1)
    for j in range(len(a1)):
        print(a1[j], s1[j])

    b_list = [i for i in range(50)]
    b0 = np.array(b_list)
    # Операции с определенным элементом массива
    b0[2] += 100
    print(b0)

    # Сложение массивов различными методами
    mas_0 = np.ones(4)
    print(mas_0, type(mas_0))
    mas_0 = np.zeros(4)
    print(mas_0, type(mas_0))
    mas_0 = np.empty(4)
    print(mas_0, type(mas_0))
    mas_0 = np.full((5, 5), 100)
    print(mas_0, type(mas_0))

    A = np.arange(0, 3000, 1)
    np.set_printoptions(threshold=sys.maxsize)
    print(A)

    # Встроенные методы в массивах
    print(A.sum())
    print(A.mean())
    print(A.min())
    print(A.max())
    print(A.std())

    a_2 = np.array([i for i in range(5)])
    b_2 = np.array([i for i in range(5, 10)])

    # Сложение срезов массивов
    c_2 = a_2[:3] + b_2[1:4]
    print(c_2)

    x_1 = np.zeros((2, 3))
    print(x_1)
    # Единичная матрица
    y_1 = np.eye(5)
    print(y_1)
    
    # Двумерны массив
    x_2 = np.array(
        [
            [i for i in range(2)],
            [i for i in range(5, 7)]
        ]
    )
    print(x_2)

    print(x_2.sum())
    print(x_2.sum(axis=0))
    print(x_2.sum(axis=1))
    print(x_2.min())
    print(x_2.mean())
    print(x_2.max())
    print(x_2.min(axis=0))
    print(x_2.min(axis=1))

    print("---")
    
    # Итерация по всем элементам двумерного массива (метод flat)
    for element in x_2.flat:
        print(element, end=" ")

    x_3 = np.array(
        [
            [i for i in range(10, 12)],
            [i for i in range(15, 17)]
        ]
    )
    print()
    print(x_2)
    print(x_3)

    # Вертикальное и горизонтальное слияние
    print(np.vstack((x_2, x_3)))
    print(np.hstack((x_2, x_3)))

    print(np.hsplit(x_3, 1))
    print(x_3[:, 1])
    
    # Матричное умножение
    print(np.dot(x_2, x_3))

    # Запись и чтение txt файлов
    np.savetxt("mas.txt", np.dot(x_2, x_3))
    r_0 = np.loadtxt("mas.txt")
    print(r_0)
```
